```python
import os
import numpy as np
patch_dir = 'D:\ShachiLab\Mukul\\Registration_matlab\\cropped'

def mse(imageA, imageB):
    # the 'Mean Squared Error' between the two images is the
    # sum of the squared difference between the two images;
    # NOTE: the two images must have the same dimension
    
    err = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
    err /= float(imageA.shape[0] * imageA.shape[1])
    
    # return the MSE, the lower the error, the more "similar"
    # the two images 
    return err
```

```python
ssim_arr = []

def calculate_ssim(gray_reg, ref):
    fil_x = 0
    fil_y = 0
    f = 100 #crop dimension f*f
    pc = 0  #patch count
    s = 20 # stride
    while fil_x + f <= 5677:
        while fil_y + f <= 7372:
            
            compare = []
            
            #get the patches from both reference and registered images
            patch_ref = ref[fil_x:fil_x + f, fil_y:fil_y + f]
            patch_reg = gray_reg[fil_x:fil_x + f, fil_y:fil_y + f]
            
            #make the directory for each pathes to save them
            dir_name = os.path.join(patch_dir, str(pc) )
            os.makedirs(dir_name)
            
            pc+=1
            compare.append(pc)
            #file name for both reference and registered images
            ref_file = os.path.join(dir_name, "patch_ref.png")
            reg_file = os.path.join(dir_name, "patch_reg.png")
            
            #save the patches
            cv2.imwrite(ref_file, patch_ref)
            cv2.imwrite(reg_file, patch_reg)
            
            #calculate the ssim value
            ssim_value = ssim(patch_ref, patch_reg)
            compare.append(ssim_value)
            
            #calculate the mse value
            mse_value =  mse(patch_ref, patch_reg)
            compare.append(mse_value)
            print(compare)
            #compare file contains both mse and ssim value corresponding to each patch
            compare_file = os.path.join(dir_name, "compare.txt")
            np.savetxt(compare_file, compare, fmt = '%1.3f')
            
            ssim_arr.append(ssim_value)
            fil_y += s
        fil_x += s
```

```python
def crop_image_pair(crop, im2, tol=0):
    # img is image data
    # tol  is tolerance
    mask = crop>tol
    mask1 = im2>tol
    return crop[np.ix_(mask.any(1),mask.any(0))][500:, :-500], im2[np.ix_(mask1.any(1),mask1.any(0))][500:, :-500]
```

```python
auto_reg = gray_reg
auto_reg, ref = crop_image_pair(auto_reg, ref)
print(ref.shape, auto_reg.shape)
```

```python
snip_size = 100
num_snip_x = int(ref.shape[0] / snip_size) # row
num_snip_y = int(ref.shape[1] / snip_size) # col
ssim_im = np.zeros(ref.shape)
ssim_vals_arr = []

def setSSIM(ref, a, ssim_im, s_x, s_y, snip_size):
    fixed = ref[s_x: min(s_x + snip_size, ref.shape[0]), s_y: min(s_y + snip_size, ref.shape[1])]
    moving = a[s_x: min(s_x + snip_size, ref.shape[0]), s_y: min(s_y + snip_size, ref.shape[1])]
    ssim_value = ssim(fixed, moving, data_range=a.max() - a.min())
    if np.mean(moving) > 20:
        ssim_vals_arr.append(ssim_value)
    ssim_im[s_x: min(s_x + snip_size, ref.shape[0]), s_y: min(s_y + snip_size, ref.shape[1])] = ssim_value
```
